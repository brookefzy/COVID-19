## 2. Simulate the susceptible-exposed-infectious- recovered (SEIR) model with fake data

*Reference: Wu, J. T., Leung, K., &
Leung, G. M. (2020). Nowcasting and forecasting the potential domestic and
international spread of the 2019-nCoV outbreak originating in Wuhan, China: a
modelling study. The Lancet.

<b>Key player:</b>
1. Susceptibles at day t: S(t) in the paper, the susceptibles is estimated by the number of inbound/outbound travellers from Wuhan
2. Exposed people at day t:E(t)
3. Infectious at day t: I(t)
4. Recoverd at day t: R(t)

<b>Other Variables: </b>
1. N: total population in a given place
2. beta: Infectious rate
3. R0: the basic reproductive number, R0=r*beta in the model below, you could replace them with the known number for each city in the paper
4. r: number of people have contacted the susceptics (scraping weibo? get a proxy)
5. r2: number of people have contacted with the exposed(scraping weibo? get a proxy)
6. a: the probability a susceptible turn to a exposed
7. y: the probability of recover

*Parameter resource: https://arxiv.org/pdf/2002.03828.pdf



```python
import matplotlib.pyplot as plt
N = 11080000     #total population for each city (district)
E = [0]       #潜伏者 Exposed people at day 1 (this is the data estimated by people coming into City from Wuhan at Day 1)
I = [641]       #传染者 Infectious people at day 1 (this is estimated with confirmed data from Hospitals)
S = [N - I[0]]#易感者 Susceptibles at day 1 (the rest of the city)
R = [0]      #康复者 Recovered at day 1 (reported from the city)


beta = 0.642 #传染概率 transmission rate
a = 1/7     #潜伏者转化为感染者概率 exposed subjects become infected

beta2 = 0.0642 #潜伏者传染正常人的概率 Get this from paper(historical data simulated) 
y = 0.135      #康复概率 (from hospital data)

T = 200

for i in range(T-1):
    susceptic=S[i]-beta*S[i]*I[i]/N-beta2*S[i]*E[i]/N
    exposed=E[i]+beta*S[i]*I[i]/N-a*E[i]+beta2*S[i]*E[i]/N
    infectious=I[i]+a*E[i]-y*I[i]
    recover=R[i]+y*I[i]
    S.append(susceptic)
    E.append(exposed)
    I.append(infectious)
    R.append(recover)



plt.plot(range(T),S,marker='', color='red', linewidth=1, alpha=0.4)

plt.plot(range(T),E,marker='', color='green', linewidth=1, alpha=0.4)
plt.plot(range(T),I,marker='', color='blue', linewidth=1, alpha=0.4)
plt.plot(range(T),R,marker='', color='orange', linewidth=1, alpha=0.4)
```

## 2.2 Exploring the Real Data, get the beta (infectious rate)

#### This experiment use the article from Beijing University, which criticize the previous one only depending on constant beta.

<b>beta=daily_new_exposed(t)/(current_infectious+ 0.1* current_exposed)</b>

##### This experiment is done at city level. But given the data sparcity, we should use the province data to get beta value

```python
dailynewconfirm=dailynewconfirm.fillna(0)
```

#### Current Infectious (t) = sum (daily confirmed cases(t)+...+daily confirmed cases(t+3))
#### Current Exposed (t) = sum (daily confirmed cases(t)+...+daily confirmed cases(t+9))

```python
delta_exposed=dailynewconfirm.shift(periods=-9)
delta_infectious=dailynewconfirm.shift(periods=-3)


exposed=dailynewconfirm.rolling(9).sum().shift(-9)
infectious=dailynewconfirm.rolling(3).sum().shift(-3)
infectious
```

```python
#put into one dataframe
infectious_df=pd.DataFrame(infectious.stack(), columns=['dailyNew_confirmed']).reset_index()
exposed_df=pd.DataFrame(exposed.stack(), columns=['dailyNew_confirmed']).reset_index()
delta_exposed_df=pd.DataFrame(delta_exposed.stack(), columns=['dailyNew_confirmed']).reset_index()
infectious_df['infected']=infectious_df['dailyNew_confirmed']
exposed_df['exposed']=exposed_df['dailyNew_confirmed']
delta_exposed_df['new_exposed']=delta_exposed_df['dailyNew_confirmed']
delta_exposed_df
```

```python
#remove noise

df=daily_frm_daily.merge(delta_exposed_df[['updateDate','new_exposed','cityName']],on=['updateDate','cityName'], how='left')
dff=df.merge(exposed_df[['updateDate','exposed','cityName']], on=['updateDate','cityName'], how='left')
seir_df=dff.merge(infectious_df[['updateDate','infected','cityName']], on=['updateDate','cityName'], how='left')
seir_df['infected']=seir_df['infected'].apply(lambda x: 0 if x<0 else x)
seir_df['exposed']=seir_df['exposed'].apply(lambda x: 0 if x<0 else x)
seir_df['new_exposed']=seir_df['new_exposed'].apply(lambda x: 0 if x<0 else x)
seir_df['beta']=seir_df['new_exposed']/(seir_df['infected']+0.1*seir_df['exposed'])


seir_df
```

```python
province_df=daily_frm[['provinceName_en','cityName_en','cityName']].drop_duplicates('cityName')
df_province=seir_df.merge(province_df,on='cityName', how='left')
df_province
```

## 2.3 Use One City Simulate the data


```python
import plotly.express as px

shanghai = provinceDFF[provinceDFF['provinceName_en']=='Shanghai'].reset_index()
fig = px.scatter(shanghai, x="updateDate", y="beta")
fig.show()
```

```python
def SEIR(currentDF, provinceName, population, daily_pop_in=0,T=40):
    N=population
    df=currentDF[currentDF['provinceName_en']==provinceName].reset_index()
    E=[df.at[0,'exposed']]
    I=[df.at[0,'infected']]
    R=[df.at[0,'cured']]
    S=[N+daily_pop_in-I[0]-E[0]-R[0]]
    beta=list(df['beta'].clip(0.06,3))
    
    a=1/7
    y=0.135
    for i in range(T-1):
        if i<9:
            b=beta[i]
        else:
            b=0
        susceptic=S[i]-b*S[i]*I[i]/N-b*0.5*S[i]*E[i]/N-R[i]
        exposed=E[i]+b*S[i]*I[i]/N-a*E[i]
        #+b*0.5*S[i]*E[i]/N
        infectious=I[i]+a*E[i]-y*I[i]
        recover=R[i]+y*I[i]
        S.append(susceptic)
        E.append(exposed)
        I.append(infectious)
        R.append(recover)
    pred=pd.DataFrame({'pre_susceptics':S,'pre_Exposed':E,'pre_Infectous':I,'pre_Recover':R})

    plt.plot(range(T),E,marker='', color='green', linewidth=1, alpha=0.4)
    plt.plot(range(T),I,marker='', color='blue', linewidth=2, alpha=0.8)
    plt.plot(range(T),R,marker='', color='orange', linewidth=1, alpha=0.4)
    plt.plot(range(len(df['infected'])),df['infected'],color='red', linewidth=2,alpha =0.8)
    return pred
    
t=SEIR(provinceDFF,'Shanghai',3000000)
```
