```python
import pandas as pd
import geopandas as gpd
import numpy as np
```

```python
#data from JhK
confirm=pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Confirmed.csv')

recover=pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Deaths.csv')
death=pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Recovered.csv')

confirm['ID']=confirm.index
recover['ID']=recover.index
death['ID']=recover.index

confirm_df=pd.DataFrame(confirm.iloc[:,4:(death.shape[1]-1)].stack()).reset_index()
recover_df=pd.DataFrame(recover.iloc[:,4:(death.shape[1]-1)].stack()).reset_index()
death_df=pd.DataFrame(death.iloc[:,4:(death.shape[1]-1)].stack()).reset_index()


confirm_df.rename(columns={'level_1':'time',0:'confirmed'}, inplace=True)
recover_df.rename(columns={'level_1':'time',0:'recover'}, inplace=True)
death_df.rename(columns={'level_1':'time',0:'death'}, inplace=True)

jhk_df=confirm_df.merge(confirm[['Province/State', 'Country/Region', 'Lat', 'Long','ID']], 
                            left_on='level_0',right_on='ID',how='left')
jhk_df['recover']=recover_df['recover']
jhk_df['death']=death_df['death']
jhk_df['time']=pd.to_datetime(jhk_df['time'])
#jhk_df.to_csv('./data/JHU_CSSE/timeseries_combined.csv')



worldDF=jhk_df.groupby(['Country/Region','time']).agg({'confirmed':'sum',
                                                           'recover':'sum',
                                                           'death':'sum'}).reset_index()

China=['Hong Kong','Taiwan','Macau','China']
#china_total=worldDF[worldDF['Country/Region'].isin(China)].reset_index()


china_all=jhk_df[jhk_df['Country/Region'].isin(China)]

jhk_df.to_csv('../../data/JHU_CSSE/timeseries_combined.csv')
china_all.to_csv('../../output/current_report.csv')
china_all
```
